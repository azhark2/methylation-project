Analysis of the "CpG density" in coding regions of Cosmic cancer genes(those genes known to be involved in cancer according to the Cosmic census) versus all other genes. The CpG density is defined as the # of cytosines in a CpG context divided by the total length of the coding region(in nucleotides) X 100.

```python
group1 = ['Non_Cancer' for i in df['CpG_Density']]
df['Gene Type'] = group1
df
```

```python
group2 = ['Cancer' for i in df2['CpG_Density']]
df2['Gene Type'] = group2
df2
```

```python
df3 = pd.concat([df, df2])
df3['CpG_Density'] = df3['CpG_Density'].apply(lambda x: x * 100)
df3 = df3.sort_values(by='CpG_Density')
df3
```

```python
import seaborn as sns, numpy as np
sns_plot = sns.violinplot( x="Gene Type", y="CpG_Density", data=df3)

# Calculate number of obs per group & median to position labels
medians = df3.groupby(['Gene Type'])['CpG_Density'].median().values
nobs = df3['Gene Type'].value_counts().values
nobs = [str(x) for x in nobs.tolist()]
nobs = ["n: " + i for i in nobs]

# Add it to the plot
pos = range(len(nobs))
for tick,label in zip(pos, sns_plot.get_xticklabels()):
    sns_plot.text(pos[tick], medians[tick] + 14.09, nobs[tick], horizontalalignment='center', size='x-large', color='r', weight='semibold')

sns_plot.set_title('CpG Density in Coding Regions of Cancer and Non-Cancer Genes')
fig = sns_plot.get_figure()
fig.savefig('/Users/khandekara2/Documents/methylationProject/03_results/CpG_density_boxplot.png')

```
